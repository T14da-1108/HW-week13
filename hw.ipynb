## FUNCTIONAL FUN

`lambda` `function composition` `function currying` `anonymous functions`

### Task

Implement the following functions. You may use anonymous functions, but it's not required.

**1. `compose`**:  
Write a function that constructs a composition of two given functions.

```python
>>> f = compose(lambda x: x ** 2, lambda x: x + 1)
>>> f(42)
1849
```

**2. `const`**:  
Create a function that, given a value, returns a constant function. This returned function should always return the initial value, regardless of its arguments.

```python
>>> f = const(42)
>>> f()
42
>>> f(range(4), range(2), foo="bar")
42
```

**3. `flip`**:  
Implement a function that takes a function `func` and returns a new function. This new function should apply `func` to its positional arguments in reverse order. Keyword arguments should be passed unchanged.

```python
>>> f = flip(map)
>>> list(f(range(5), range(0, 10, 2), lambda x, y: x**y))
[1, 2, 16, 216, 4096]
>>> list(map(lambda x, y: x**y, range(0, 10, 2), range(5)))
[1, 2, 16, 216, 4096]
```

**4. `curry`**:  
Write a function that takes a function `func` and returns a new function with some positional and keyword arguments fixed. These fixed arguments should be passed to `curry` alongside `func`.

```python
>>> def pow(x: int, p: int) -> int:
...     return x**p
...
>>> ten_pow = curry(pow, 10)
>>> ten_pow(3)
1000
>>> square = curry(pow, p=2)
>>> square(42)
1764
```

### About the Task

These tasks cover a variety of functional programming concepts including function composition and currying. These are important techniques that can lead to cleaner, more readable code, and are commonly used in functional programming languages like Haskell. They can also come up in technical interviews.
